```python
import pandas as pd
```

```python
df = pd.read_csv("experiment_data.csv")
df
```

```python
df_agg = (df.groupby(["Shape"])
             .mean(numeric_only=True))
df_agg
```

```python
cols = ["Total layout time (s)", "Components", "Sparsity (%)", "Bounding box (cells)", "Blank orthoconvex hull cells", "Average set squareness", "Average set vertices", "Average set straight sides"]
```

```python
rows = []

for dataset, row in df_agg.iterrows():
    line = [dataset]

    for col in cols:
        a = row[col]
        line.extend([f"{a:.3g}"])

    rows.append(" & ".join(line) + r" \\")

print("\n".join(rows))
```
