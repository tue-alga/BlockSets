```python
import pandas as pd
```

```python
df = pd.read_csv("experiment_data.csv")
df
```

```python
df = df[df["Components"] > 1]
df
```

```python
cols = ["Components", "Duplicated sets", "Set copies",
        "Max element size difference", "Sparsity (%)", "Bounding box (cells)", "Total split time (s)"]
df_agg = (df.groupby(["Dataset", "Split"])
             .mean(numeric_only=True)[cols]
             .unstack("Split"))
df_agg = df_agg[df_agg["Components", "RectEuler"].notnull()]
df_agg
```

```python
def color_pair(a, b):
    if pd.isna(a) or pd.isna(b):
        return str(a), str(b)
    if a < b:
        return f"\\good{{{a:.3g}}}", f"\\bad{{{b:.3g}}}"
    elif a > b:
        return f"\\bad{{{a:.3g}}}", f"\\good{{{b:.3g}}}"
    else:
        return f"\\equal{{{a:.3g}}}", f"\\equal{{{b:.3g}}}"
```

```python
rows = []

for dataset, row in df_agg.iterrows():
    line = [dataset]

    for col in cols:
        a = row[(col, "BlockSets")]
        b = row[(col, "RectEuler")]

        ca, cb = color_pair(a, b)
        line.extend([ca, cb])

    rows.append(" & ".join(line) + r" \\")

print("\n".join(rows))
```
